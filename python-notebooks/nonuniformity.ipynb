```python
import numpy as np 
import matplotlib.pyplot as plt
import imageio as io
```

```python
def generate_noise(size, lowpass):
    W = np.random.random((size, size))
    freqs = np.fft.fftfreq(size)
    f = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            f[i,j] - freqs[i]**2 + freqs[j]**2
    
    fW = np.fft.fft2(W) / (f + 0.00000001)
    filter = np.zeros((size, size))
    filter[0:lowpass, 0:lowpass] = 1
    fW = fW * filter
    noise = np.abs(np.fft.ifft2(fW))
    noise = (noise - np.min(noise)) / (np.max(noise) - np.min(noise))
    return noise

R = (np.ones((500,500)) * 0.6 ) + 0.4 * generate_noise(500,3)
G = (np.ones((500,500)) * 0.6 ) + 0.4 * generate_noise(500,3)
B = (np.ones((500,500)) * 0.6 ) + 0.4 * generate_noise(500,3)

img = np.dstack((R,G,B))

plt.imshow(img)

print(img.dtype)
io.imwrite("./non-uni-ex1.png",(img*255.0).astype(np.uint8), format="png")
```

```python
eff = np.ones((500,500))

for i in range(500):
    for j in range(500):
        x = ((i - 250) / 500) * 15
        y = ((j - 250) / 500) * 15

        dist_deg = np.sqrt(y**2 + x**2)

        eff[i][j] = eff[i][j] * np.cos(dist_deg * np.pi/180)**100

plt.imshow(eff)
plt.show()


fig = plt.figure()
ax = fig.add_subplot()
plt.plot(np.linspace(-15,15, 500), eff[250:251,:][0])
 
# square plot
ax.set_aspect(1.0/ax.get_data_ratio(), adjustable='box')
plt.show()

plt.imshow(np.dstack((eff,eff,eff)) * img)
plt.show()

imgg = np.dstack((eff,eff,eff)) * img

io.imwrite("./non-uni-ex-2.png",(imgg*255.0).astype(np.uint8), format="png")
        
```
